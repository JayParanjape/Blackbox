```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def make_plot_and_print_tr_progress(path):
    with open(path, 'r') as f:
        lines = f.readlines()
    tr_loss = []
    tr_dice = []
    tr_epochs = []

    for i,s in enumerate(lines):
        tr_epochs.append(i)
        if 'Average loss on the tr set' in s:
            loss = float(s[s.find('(')+1:s.find(')')])
            tr_loss.append(loss)
        if 'Average dice on the tr set' in s:
            loss = float(s[s.find('(')+1:s.find(')')])
            tr_dice.append(loss)
    
    return tr_loss, tr_dice, tr_epochs
```

```python

loss_m1, dice_m1, epoch_m1 = make_plot_and_print_tr_progress('saved_models_and_configs/isic/model1_medsam/training_progress.log')
loss_vp, dice_vp, epoch_vp = make_plot_and_print_tr_progress('saved_models_and_configs/isic/vp_medsam/training_progress.log')
loss_spsa, dice_spsa, epoch_spsa = make_plot_and_print_tr_progress('saved_models_and_configs/isic/model1_spsa_nogc/training_progress.log')
```

```python
plt.plot(loss_m1[0::3][:30])
plt.plot(loss_vp[0::3][:30])
# plt.plot(loss_spsa)
plt.legend(['BAPS','VPT'])
plt.ylabel('Average Loss')
plt.xlabel('Iterations')
plt.show()
```
